**Scatterplot**

```python
df.plot(kind='scatter', x='turns', y='winner')
plt.title('Number of Turns by Colour of Winner')
```

A few things that I find interesting about this is: 
- It seems that the threshold of turns increases when there is a draw. This must be because the more turns you take,and the more pieces that are taken out over time, the less you are able to do or possibly win.
- I also find it interesting that both the white winners and black winners sections are almost the same. I thought that by chance, one might win with less turns, or vice versa.

**Histogram**

```python
df['turns'].plot(kind='hist', bins=50, figsize=(15,5), facecolor='green',edgecolor='black')
plt.title('Number of Turns Each Game')
```

A few things I find interesting about this histogram: 
- The amount of games durastically increased around 40 and fell quite faster around 75. This might be quite common, it is surprising to me because I have never seen data about the average of turns in a chess match. 
- I am also curious about how many games near the end of the spectrum ended in a draw.

**The Minimum Number of Turns vs. The Maximum Number of Turns**

```python
column = df["turns"]
min_value = column.min()
print("The minimum number of turns is:",min_value)
column = df["turns"]
max_value = column.max()
print("The maximum number of turns is:",max_value)
```

I find the minimum number of moves very interesting because I cannot figure out how someone would win a chess match in only one move. Frankly, I thought it was impossible. This might be an error in the data, but if not, winning in one move is incredible.
